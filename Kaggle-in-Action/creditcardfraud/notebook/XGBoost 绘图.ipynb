```python
import xgboost as xgb
import pandas as pd

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
data['feature_names']
```

```python
# columns name can not have blank if you want to draw tree with feature name
data['feature_names'] = [
    ''.join(
        [character if character != ' ' else '_' for character in featureName])
    for featureName in data['feature_names']
]
data['feature_names']
```

```python
data['feature_names'][0]
```

```python
data.data
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df
```

```python
from xgboost import XGBClassifier

clf = XGBClassifier()
clf.fit(df, data['target'])
```

```python
xgb.to_graphviz(clf)
```
