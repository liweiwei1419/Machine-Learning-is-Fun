```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic.info()
```

```python
# sns.set(style="darkgrid", palette="muted", color_codes=True)
sns.distplot(titanic[titanic['age'].notnull()]['age'])
plt.show()
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
sns.distplot(titanic['age'])
plt.show()
```

```python
sns.countplot(x="embarked",data=titanic)
plt.show()
```

```python
titanic['embarked'] = titanic['embarked'].fillna('S')
titanic.isnull().sum()
```

```python
titanic.head()
```

```python
titanic = titanic.drop(
    [
        'survived', 'pclass', 'sibsp', 'parch', 'who', 'adult_male', 'deck',
        'embark_town'
    ],
    axis=1)
titanic.head()
```

```python
sns.countplot(x="sex",data=titanic)
plt.show()
```

```python
# x  y 
sns.boxplot(x='sex', y='age', data=titanic)
plt.show()
```

```python
sns.countplot(x="class", data=titanic)
plt.show()
```

```python
sns.violinplot(x="class", y="age", data=titanic)
plt.show()
```

```python
sns.countplot(x="alone", data=titanic)
plt.show()
```

```python
sns.countplot(x="alive", data=titanic)
plt.show()
```

```python
# hue 
sns.countplot(x='alive', hue='sex', data=titanic)
plt.show()
```

```python
# FacetGrid 
g = sns.FacetGrid(titanic, col='sex')
g.map(sns.countplot, 'alive')
```

```python
def agelevel(age):
    if age <= 16:
        return 'child'
    elif age >= 60:
        return 'aged'
    else:
        return 'midlife'


titanic['age_level'] = titanic['age'].map(agelevel)
```

```python
titanic.head()
```

```python
sns.countplot(x='age_level', data=titanic)
plt.show()
```

```python
sns.countplot(x='alive', hue='age_level', data=titanic)
plt.show()
```

```python
g = sns.FacetGrid(titanic, col='class', row='alone')
g.map(sns.countplot, 'alive')
plt.show()
```
